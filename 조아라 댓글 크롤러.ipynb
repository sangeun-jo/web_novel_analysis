```python
from bs4 import BeautifulSoup
from selenium import webdriver
import re
import time
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from web_novel_analysis.base import get_tags
from web_novel_analysis.base import wordcloud

def joara_comment(url):
    driver = webdriver.Chrome('D:/chromedriver.exe')
    driver.get(url)
    commentList = []
    while True:  
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        MoreBtn = soup.select_one('#commentMoreBtn')
        if MoreBtn:
            driver.find_element_by_xpath('//*[@id="commentMoreBtn"]/a').click()
            time.sleep(0.3)
        else:
            break; 
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    commentList = soup.find_all('p', class_='comment')
    c_list = []
    for co in commentList:
        c_list.append(co.text)
    return c_list
```

```python
keyword = get_tags(joara_comment('http://www.joara.com/romancebl/view/book_intro.html?book_code=1448720&refer_type=&book_dtype=comment'))
```
