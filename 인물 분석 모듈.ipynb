```python
import time
def ma_contents(contents):
    ma_results = []
    ma = Komoran()
    for content in contents:
        try:
            ma_by_content = ma.pos(content)
        except:
            continue
        ma_results.append(ma_by_content)
    return ma_results

```

```python
f = open("[뾰족가시] 감금(미완).txt", 'r', encoding='UTF-8')
lines = f.readlines()
f.close()
```

```python
ma_results = ma_contents(lines)
```

```python
FILTER_IN_POSES=['NNG', 'NNP']
STOPWORD=['으로', '어요']

def ma_results_select(ma_results, FILTER_INPOSES, STOPWORD):
    
    ma_results_selected = []  
    
    for ma_result in ma_results:
        ma_selected=[]
        for morph_lex, morph_cat in ma_result:
            if morph_lex not in STOPWORD and morph_cat in FILTER_INPOSES:
                ma_selected.append(morph_lex)
                
        if len(ma_selected)>0:
            ma_results_selected.append(ma_selected)
                
    return ma_results_selected
ma_results_selected = ma_results_select(ma_results, FILTER_IN_POSES, STOPWORD)
```

```python
from collections import Counter
from itertools import combinations
```

```python
def word_cooc_matrics(contents):
    
    word_cooc_mat = Counter()
    for line in contents:
        for word1, word2 in combinations(line, 2):
            if len(word1) == 1 or len(word2) == 1:
                continue
            if word1 == word2:
                continue
            elif word_cooc_mat[(word2, word1)] >=1:
                word_cooc_mat[(word2, word1)] +=1
            else:
                word_cooc_mat[(word1, word2)] += 1
    return word_cooc_mat
```

```python
def sna_graph(word_coocs, NETWORK_MAX):
    
    G = nx.Graph()
    i = 0
    
    word_cooc_mat = []
    for word, count in word_coocs.most_common():
        w1, w2 = word
        word_cooc_mat.append((w1, w2, count)
    
    # edge 를 생성한다
    for word1, word2, count in word_cooc_mat:
        i +=1
        if i > NETWORK_MAX: break
        G.add_edge(word1, word2, weight=count)
       
    # MST 모델을 만든다. 
    T = nx.minimum_spanning_tree(G)
    nodes = nx.nodes(T)
    degrees = nx.degree(T)
    
    # node 사이즈를 정해준다
    node_size = []
    for node in nodes:
        ns = degrees[node]*100
        node_size.append(ns)

    # 폰트 정리
    if sys.platform in ["win32", "win64"]:
        font_name = "malgun gothic"
    elif sys.platform == "darwin":
        font_name = "AppleGothic"
        
    # 분석결과를 화면에 표시한다
    plt.figure(figsize=(16,12))    
    nx.draw_networkx(T,
           pos=nx.fruchterman_reingold_layout(G, k=0.5),
           node_size=node_size,
           node_color="yellow",
           font_family=font_name,
           label_pos=1, #0=head, 0.5=center, 1=tail
            with_labels=True,
            font_size=12 )

    plt.axis("off")
    plt.show()
```
