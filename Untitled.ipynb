```python
from web_novel_analysis.base import get_tags
from konlpy.tag import Twitter
from collections import Counter
import re
```

```python
qs = TodayBest.objects.all()
```

```python
kor_eng = re.compile('[^ a-zA-Zㄱ-ㅣ가-힣]+') #한글, 영어만 추출 
txt = ''
for nov in qs:
    txt = txt + ' ' + nov.title + ' ' + nov.intro
cleaned_text = kor_eng.sub(' ', txt)
spliter = Twitter()
nouns = spliter.nouns(cleaned_text)
nouns = [n for n in nouns if len(n) > 1] #한글자 단어 삭제 
count = Counter(nouns)
return_list = {}
for n, c in count.most_common(20):
    if n == '표지' or n == '소설':
        continue
    return_list[n] = c
```
